```python
import time
import ast
from lxml import html
import requests
```

```python
from selenium import webdriver
```

```python
def extract_info(offer_link):
    
    driver.get(offer_link)
    ## Getting location
    location = driver.find_element_by_xpath('//a[contains(@href, "#map")]').get_attribute("text")
    
    ## Getting total price
    total_price = driver.find_element_by_xpath("//div[@class='css-1vr19r7']").text
    
    ## Getting property info 
    property_info = driver.find_elements_by_xpath("//div[@class='css-1ci0qpi']//li")
    property_info = [info.text for info in property_info]
    property_info = [info.split(": ") for info in property_info]
    
    
    ## Getting date of an offer
    date_info = driver.find_element_by_xpath("//div[@class='css-lh1bxu']")
    date_info = date_info.text
    results = (location, total_price, date_info, property_info)
    
    with open("results.txt", "a", encoding = "utf-8") as file:
        file.write(str(results))
        file.write("\n")
    
#     return results
```

```python
driver = webdriver.Chrome()
all_links_of_properties = []

for page_number in range(1, 500):
    print(page_number)
    driver.get("https://www.otodom.pl/sprzedaz/mieszkanie/warszawa/?page={}".format(page_number))
    property_links = driver.find_elements_by_xpath('//h3//a[contains(@href,"otodom")]')
    property_links = [link.get_attribute('href') for link in property_links]
    all_links_of_properties.append(property_links.copy())
    time.sleep(10)
    
with open("links.txt", "a", encoding = "utf-8") as file:
    file.write(str(all_links_of_properties))
#         file.write("\n")
```

```python
final_links = [x for y in all_links_of_properties for x in y]
```

```python
all_res = []
```

```python
def xml_parser(link):
    global all_res
    page = requests.get(link)
    tree = html.fromstring(page.content)
    
    location = tree.xpath('//a[contains(@href, "#map")]/text()')
    location = location[0]
    
    ## Getting total price
    total_price = tree.xpath("//div[@class='css-1vr19r7']/text()")
    total_price = total_price[0]
    
    ## Getting property info 
    property_info = tree.xpath("//div[@class='css-1ci0qpi']//li")
    property_info = [info.text_content() for info in property_info]
    property_info = [info.split(": ") for info in property_info]


    ## Getting date of an offer
    date_info = tree.xpath("//div[@class='css-lh1bxu']/text()")
    results = (location, total_price, date_info, property_info, link)
    
    with open("results.txt", "a", encoding = "utf-8") as file:
        file.write(str(results))
        file.write("\n")
    
    
    
    time.sleep(3)
```

```python
control = []
for num, link in enumerate(final_links):
    
    if num % 100 == 0:
        print (num)
    
    try:
        xml_parser(link)
        control = []
    except:
        control.append(num)
        pass
    
    
    if len(control) == 20:
        print(num, "too many errors in a row")
        break
    
```
